```python
from neko.core.network import Network
from neko._visual.visualize_network import NetworkVisualizer
from neko.inputs import Universe
import pandas as pd
```

```python
data_folder = "../data/"

figures_folder = "../Figures/"

sif_output = "../sif_files/"
```

```python
Huri = pd.read_csv('../data/HURI_full_data_trasnlated.csv', usecols=["source_UniProtKB-Swiss-Prot", "target_UniProtKB-Swiss-Prot"])
```

```python
resources = Universe()
mapping = {"source_UniProtKB-Swiss-Prot": "source", "target_UniProtKB-Swiss-Prot": "target"}
resources.add_resources(Huri, columns=mapping,  reset_index=True)
resources.build()
```

```python
df = pd.read_csv("../../data/Northcott_Lin_genes.csv")
```

```python
import py4cytoscape as p4c
```

```python
# Connect to Cytoscape
p4c.cytoscape_ping()  # This will check if Cytoscape is running
```

```python
missing_nodes = [
    "ARID2", "EPHA7", "DLX4", "EN1", "FOXF1", "HOMEZ", "HOXD11", 
    "HOXD13", "HOXD9", "IRX5", "NKX6-1", "ONECUT2", "SHOX2", 
    "EPHA7", "IDH1", "ATOH1", "BBX", "DMRTA1", "POU3F3", 
    "KBTBD4", "KBTBD4", "ARID5B", "BARHL1", "CUX2", "DLX5", 
    "EN2", "LHX5", "UNCX"
]
```

```python
wnt_list_updated = set(wnt_genes) - set(missing_nodes)
```

```python
# Your list of nodes to be colored
node_list = list(wnt_list_updated)  # Replace with your actual node IDs
```

```python
# Get the node names in the current network
node_table = p4c.get_table_columns('node', columns=['name'])

# Extract the list of node names
cytoscape_nodes = node_table['name'].tolist()

# Print or inspect the node names
print(cytoscape_nodes)
```

```python
len(cytoscape_nodes)
```

```python
def filter_existing_nodes(group_nodes):
    updated_list = []
    for gene in group_nodes:
        if gene in cytoscape_nodes:
            updated_list.append(gene)

    return updated_list
```

```python
p4c.set_node_color_bypass(filter_existing_nodes(node_list), '#FFF54E')  # Replace with your desired color
```

```python
p4c.set_node_color_bypass(filter_existing_nodes(shh_genes), '#B2E782')  # Replace with your desired color
```

```python
p4c.set_node_color_bypass(filter_existing_nodes(g3_genes), '#FED303')  # Replace with your desired color
```

```python
p4c.set_node_color_bypass(filter_existing_nodes(g4_genes), '#C37578')  # Replace with your desired color
```

```python
p4c.create_group("WNT", filter_existing_nodes(node_list), nodes_by_col="name")
```

```python
p4c.create_group("SHH", filter_existing_nodes(shh_genes), nodes_by_col="name")
```
